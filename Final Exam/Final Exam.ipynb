```python
#Without Magnets (Velocity and Acceleration)
start_A1 = np.array([2.665603, 1.7601, 0.703872, 1.449586, 1.356752])
stop_A1 = np.array([2.675778, 1.768268, 0.711501, 1.456485, 1.363213])
start_A2 = np.array([2.686363, 1.776764, 0.719542, 1.463848, 1.369698])
stop_A2 = np.array([2.696329, 1.784748, 0.727161, 1.470657, 1.376041])

start_B1 = np.array([3.136701, 2.130822, 1.067512, 1.774345, 1.651547])
stop_B1 = np.array([3.144863, 2.137245, 1.07347, 1.77968, 1.656338])
start_B2 = np.array([3.152199, 2.1429, 1.078775, 1.78441, 1.660671])
stop_B2 = np.array([3.160385, 2.149228, 1.084711, 1.789716, 1.66544])

t_A1 = np.subtract(stop_A1, start_A1)
t_A2 = np.subtract(stop_A2, start_A2)
t_B1 = np.subtract(stop_B1, start_B1)
t_B2 = np.subtract(stop_B2, start_B2)

v_A1 = np.divide(0.01, t_A1)
v_A2 = np.divide(0.01, t_A2)
v_B1 = np.divide(0.01, t_B1)
v_B2 = np.divide(0.01, t_B2)

a_A = np.abs(np.divide(np.subtract(v_B2, v_A1), np.subtract(start_B2, start_A1)))
a_B = np.abs(np.divide(np.subtract(v_B2, v_B1), np.subtract(start_B2, start_B1)))

error_A = 0.1*a_A
error_B = 0.1*a_B

sintheta = np.array([0.091, 0.127, 0.162, 0.199, 0.220])

p, q = np.polyfit(sintheta, a_A, 1)

plt.errorbar(sintheta, a_A, yerr=error_A, fmt='o', markersize=5, ecolor='black')
plt.plot(sintheta, p*sintheta+q, color='red')
plt.xlabel(r'$\sin(\theta)$')
plt.ylabel(r'Acceleration [m/s$^2$]')
plt.title('Acceleration-Time for Rider without Magnets')
plt.text(0.1, 1.8, r'$g=8.89 \ m/s^2$')
print(p)
#plt.errorbar(sintheta, a_B, yerr=error_B, fmt='o', markersize=5, ecolor='black')
plt.show()
```
