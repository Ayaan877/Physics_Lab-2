```python
A = []

for i in range(len(T_A)):
    
    a, b, c = np.polyfit(T_A[i][20:], Y_A[i][20:], 2)
    
    A.append(2*a)
    
a = np.array(A)

p, q = np.polyfit(m_A, a, 1)

m_error = np.array([0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001])

print("Moment of Interia (I) =", np.round((-9.81*0.0075**2)/p, 5), "kgm^2")
print("Frictional Torque (T) =", np.round(q*((-9.81*0.0075**2)/p)/0.0075, 5), "Nm")

plt.errorbar(m_A, a, xerr=m_error, fmt='o', ecolor='black')
plt.plot(m_A, p*m_A+q, color='red')
plt.xlabel(r'Mass [kg]')
plt.ylabel(r'Acceleration [m/s$^2$]')
plt.show()
```
