```python
#Part A (Acceleration Due to Gravity)
T_A = np.array([t1, t2, t3, t4, t5, t6, t7, t8, t9, t10, t11, t12, t13, t14, t15, t16, t17, t18, t19, t20, t21, t22, 
                t23], dtype=object)
Y_A = np.array([y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11, y12, y13, y14, y15, y16, y17, y18, y19, y20, y21, y22,
               y23], dtype=object)
```

```python
#Acceleration of TT Ball
A = []

for i in range(len(T_A)):
    
    p, q, r = np.polyfit(T_A[i], Y_A[i], 2)
    
    A.append(2*p)
    
a = np.array(A)
```

```python
#Histogram and Gaussian Fit 
def gauss(x, amp, mu, sigma):
    return amp*np.exp(-(x-mu)**2/(2*sigma**2))

counts, bins, patches = plt.hist(a, bins=8)

b = []

for i in range(len(bins[:-1])):
    B = bins[:-1][i]-np.diff(bins)[i]
    
    b.append(B)
    
b = np.array(b)

par, cov = curve_fit(gauss, b, counts, p0=[5, -9.629348546055695, 0.223409740002056])

x = np.linspace(-10.2, -9.2, 1000)

plt.plot(x, gauss(x, *par), color='red')
plt.vlines(par[1], 0, 5.2, linestyle='--', color='orange')

print('Mean =', par[1])
print('Standard Deviation =', par[2])

plt.xlabel(r'Acceleration [m/s$^2$]')
plt.ylabel('Frequency')
plt.text(-9.7, 5, r'$g = -9.75 \ m/s^2$')
plt.show()
```

```python
#Part B (Terminal Velocity)
T_B = np.array([t_1, t_2, t_3, t_4, t_5, t_6, t_7, t_8, t_9, t_10, t_11, t_12, t_13, t_14, t_15, t_16], dtype=object)
Y_B = np.array([y_1, y_2, y_3, y_4, y_5, y_6, y_7, y_8, y_9, y_10, y_11, y_12, y_13, y_14, y_15, y_16], dtype=object)
m = np.array([0.8, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2, 2.3, 2.4])
V = []

for i in range(len(T_B)):
        v = np.divide(np.diff(Y_B[i]), np.diff(T_B[i]))
        
        V.append(v)
        
V_B = np.array(V, dtype='object')
```

```python
#Tan Hyperbolic Fit

G = []
V_t = []

for i in range(len(V_B)):
    def tanh(x, v_t, g):
        return v_t*np.tanh(g*x/v_t)
    
    x = np.linspace(T_B[i][0], T_B[i][-1], 1000)

    par, cov = curve_fit(tanh, T_B[i][1:], V_B[i], p0=[1.6, -9.81])
    
    G.append(par[1])
    V_t.append(par[0])
    
    plt.hlines(-par[0], T_B[i][0], T_B[i][-1], linestyle='--', color='orange')
    plt.plot(x, tanh(x, *par), color='red')
    plt.scatter(T_B[i][1:], V_B[i])
    plt.xlabel('Time [s]')
    plt.ylabel('Velocity [m/s]')
    print('g =', par[1])
    print('v_t =', -par[0])
    plt.text(0.3, -0.5, r'$v_t =$'+str(np.round(-par[0], 3)), fontsize=35)
    plt.text(0.3, -1, r'$g =$'+str(np.round(par[1], 3)), fontsize=35)
    plt.show()
```

```python
#Damped Free Fall Histogram

def gauss(x, amp, mu, sigma):
    return amp*np.exp(-(x-mu)**2/(2*sigma**2))

counts, bins, patches = plt.hist(G, bins=10)

q = []

for i in range(len(bins[:-1])):
    B = bins[:-1][i]-np.diff(bins)[i]
    
    q.append(B)
    
P = np.array(q)

par, cov = curve_fit(gauss, P, counts, p0=[5, -8.251194751273681, 1.1381482132278145])

x = np.linspace(-11.5, -6.5, 1000)

plt.plot(x, gauss(x, *par), color='red')
plt.vlines(par[1], 0, 4, linestyle='--', color='orange')

print('Mean =', par[1])
print('Standard Deviation =', par[2])

plt.xlabel(r'Acceleration [m/s$^2$]')
plt.ylabel('Frequency')
plt.text(-9.8, 4, r'$g = -8.47 \ m/s^2$')
plt.show()
```

```python
#Terminal Velocity and Mass

p, q = np.polyfit(np.log(m), np.log(V_t), 1)

m_error = []

for i in range(len(m)):
    M = np.abs(np.abs(np.log(m[i]+0.0625))-np.abs(np.log(m[i]-0.0625)))
    
    m_error.append(M)

m_error = np.array(m_error)

x = np.linspace(-0.27, 0.95, 100)

plt.errorbar(np.log(m), np.log(V_t), xerr=m_error, fmt='o', ecolor='black')
plt.plot(x, p*x+q, color='red')
plt.xlabel('Mass [g]')
plt.ylabel('Terminal Velocity [m/s]')
print('Slope =', p)
plt.show()
```
